```python
import pandas as pd
import numpy as np
```

```python
# ==========================
# 0. パラメータ
# ==========================
EXCEL_PATH = "echizen_companies_40_filled (2).xlsx"
SHEET_NAME = "Sheet1"          # 必要に応じて変更
EDGE_COUNT = 125               # 欲しいエッジ本数（元のBaseと合わせる）
ALPHA = 0.5                    # 規模の影響度（重力モデルの指数）
```

```python
# ==========================
# 1. データ読み込み
# ==========================
df = pd.read_excel("/content/echizen_companies_40_filled (2).xlsx", sheet_name=SHEET_NAME)

# ここはあなたの列名に合わせて調整
ID_COL = "ID"
NAME_COL = "企業名"
IND_COL = "業種"
EMP_COL = "従業員数"

# 従業員数を数値化 & 欠損は中央値で補完
df[EMP_COL] = pd.to_numeric(df[EMP_COL], errors="coerce")
df[EMP_COL] = df[EMP_COL].fillna(df[EMP_COL].median())
```

```python
# ==========================
# 2. 産業セクタ & 川上/川中/川下 tier を付与
#    （DBJ 越前市図を参考にしたルール）
# ==========================

def classify_tier(industry: str) -> str:
    """業種名から川上(up) / 川中(mid) / 川下(down) をざっくり割り当てる."""
    s = str(industry)

    # 川下：食品・小売・サービス・輸送用機械（完成品寄り）
    if any(k in s for k in ["食品", "飲食", "小売", "卸", "不動産", "輸送用機械"]):
        return "down"

    # 川中：建設・設備・金属加工・機械・電子部品など
    if any(k in s for k in ["建設", "木材", "環境設備", "金属加工", "機械部品", "電気", "電子"]):
        return "mid"

    # 川上：樹脂・化学・材料・繊維など
    if any(k in s for k in ["樹脂", "化学", "材料", "繊維", "眼鏡"]):
        return "up"

    # よく分からないものはとりあえず川中にしておく（中間財が多い想定）
    return "mid"


# セクタは今回は業種そのものを使う（必要ならここで正規化してもOK）
df["sector"] = df[IND_COL].astype(str)
df["tier"] = df[IND_COL].apply(classify_tier)
```

```python
# ==========================
# 3. 企業規模 size_i を計算（重力モデルのM_i）
# ==========================
df["size"] = np.log1p(df[EMP_COL].values.astype(float))  # log(従業員数+1)
```

```python
# ==========================
# 4. 産業間補完度 S(s_i, s_j) を定義
#    - tier 間の基本スコア（川上→川中→川下）
#    - 特定産業ペアのブースト（材料→電子→自動車部品など）
# ==========================

# 川上/中/下の基本スコア（DBJ図の「階層構造」を意識）
TIER_BASE = {
    ("up", "mid"): 3.0,
    ("mid", "up"): 3.0,
    ("mid", "down"): 3.0,
    ("down", "mid"): 3.0,
    ("up", "down"): 2.0,
    ("down", "up"): 2.0,
    ("up", "up"): 0.8,
    ("mid", "mid"): 1.0,
    ("down", "down"): 0.8,
}

# 産業名から、より粗い「産業グループ」を抽出
def sector_group(industry: str) -> str:
    s = str(industry)
    if any(k in s for k in ["樹脂", "化学", "材料"]):
        return "materials"
    if any(k in s for k in ["繊維", "眼鏡"]):
        return "textile"
    if any(k in s for k in ["金属加工", "鋼材"]):
        return "metal"
    if any(k in s for k in ["電気", "電子"]):
        return "electronics"
    if "輸送用機械" in s or "自動車部品" in s:
        return "auto"
    if "食品" in s or "飲食" in s:
        return "food"
    if any(k in s for k in ["小売", "卸", "不動産"]):
        return "trade"
    if "建設" in s or "木材" in s or "環境設備" in s:
        return "construction"
    return "other"

df["sector_group"] = df["sector"].apply(sector_group)

# グループ間の追加ブースト（DBJ図で太そうなところ）
def sector_boost(g1: str, g2: str) -> float:
    pair = {g1, g2}

    # 材料 <-> 電子部品
    if pair == {"materials", "electronics"}:
        return 1.4

    # 電子部品 <-> 自動車部品
    if pair == {"electronics", "auto"}:
        return 1.5

    # 金属加工 <-> 自動車部品
    if pair == {"metal", "auto"}:
        return 1.4

    # 建設 <-> 金属/材料（建材）
    if pair == {"construction", "metal"} or pair == {"construction", "materials"}:
        return 1.3

    # 食品 <-> trade（卸売・小売）
    if pair == {"food", "trade"}:
        return 1.4

    # 繊維 <-> trade
    if pair == {"textile", "trade"}:
        return 1.3

    # それ以外はブーストなし
    return 1.0


def S(i: int, j: int) -> float:
    """企業 i, j の産業間補完度 S(s_i, s_j) を計算."""
    ti, tj = df.loc[i, "tier"], df.loc[j, "tier"]
    base = TIER_BASE.get((ti, tj), 0.5)  # 未定義は弱めに 0.5

    g1, g2 = df.loc[i, "sector_group"], df.loc[j, "sector_group"]
    boost = sector_boost(g1, g2)

    return base * boost
```

```python
# ==========================
# 5. 全ペアの重力スコア w_ij を計算
# ==========================
n = len(df)
scores = []

for i in range(n):
    for j in range(i + 1, n):   # 無向グラフを想定して i<j のみ
        sij = S(i, j)
        if sij <= 0:
            continue

        si = df.loc[i, "size"]
        sj = df.loc[j, "size"]
        size_factor = (si * sj) ** ALPHA  # gravity-model: (M_i M_j)^alpha

        w_ij = sij * size_factor
        scores.append((i, j, w_ij))
```

```python
# ==========================
# 6. 上位 EDGE_COUNT 本をエッジとして採用
# ==========================
scores.sort(key=lambda x: x[2], reverse=True)

E = min(EDGE_COUNT, len(scores))
selected = scores[:E]

edges = [(i, j) for (i, j, w) in selected]
print(f"選ばれたエッジ本数: {len(edges)}")
```

```python
# ==========================
# 7. PyTorch Geometric 用 edge_index (2, E) に変換して保存
# ==========================
edge_index = np.array(edges).T  # shape = (2, E)
np.save("edge_index_base_gravity.npy", edge_index)

# ノード情報も tier や sector_group 付きで保存しておくと便利
node_cols = [ID_COL, NAME_COL, IND_COL, "sector_group", "tier", EMP_COL, "size"]
df[node_cols].to_csv("nodes_echizen_with_tier.csv", index=False)

print("edge_index_base_gravity.npy と nodes_echizen_with_tier.csv を保存しました。")
```

```python
# =========================
# 1. ファイル読み込み
# =========================
nodes = pd.read_csv("nodes_echizen_with_tier.csv")
edge_index_np = np.load("edge_index_base_gravity.npy")   # shape = (2, E)

print(nodes.head())
print("edge_index shape:", edge_index_np.shape)

# =========================
# 2. カテゴリ特徴量の one-hot 化
#    - sector_group
#    - tier（up / mid / down）
# =========================

cat_cols = ["sector_group", "tier"]

enc = OneHotEncoder(sparse_output=False)
cat_ohe = enc.fit_transform(nodes[cat_cols])   # shape = (N, C)

print("one-hot 次元:", cat_ohe.shape[1])
print("カテゴリ値:", enc.get_feature_names_out(cat_cols))

# =========================
# 3. 連続特徴量（size）を標準化
#    size = log(従業員数+1) はすでに入っている想定
# =========================

cont_cols = ["size"]
cont_vals = nodes[cont_cols].values.astype(float)   # shape = (N, 1)

scaler = StandardScaler()
cont_scaled = scaler.fit_transform(cont_vals)       # shape = (N, 1)

print("連続特徴 shape:", cont_scaled.shape)

# =========================
# 4. 特徴量を結合して X を作成
# =========================

X = np.hstack([cat_ohe, cont_scaled])   # (N, C + 1)
print("X shape:", X.shape)

# torch tensor に変換
x = torch.tensor(X, dtype=torch.float)

# edge_index を PyG 形式に
edge_index = torch.tensor(edge_index_np, dtype=torch.long)

print("x:", x.shape, "edge_index:", edge_index.shape)

# =========================
# 5. PyTorch Geometric の Data オブジェクトにまとめる
# =========================

from torch_geometric.data import Data

data = Data(x=x, edge_index=edge_index)

print(data)
```

```python
from torch_geometric.transforms import RandomLinkSplit
```

```python
# train / val / test に分割（正例＋負例も作る）
transform = RandomLinkSplit(
    num_val=0.1,                 # 検証用 正例の割合
    num_test=0.1,                # テスト用 正例の割合
    is_undirected=True,          # 無向グラフ
    add_negative_train_samples=True,  # train に負例も自動で作る
)

train_data, val_data, test_data = transform(data)

print(train_data)
print("train edge_index:", train_data.edge_index.shape)
print("train edge_label:", train_data.edge_label.shape)        # +1/-1 (正例/負例)
print("val edge_label:", val_data.edge_label.shape)
print("test edge_label:", test_data.edge_label.shape)
```

```python
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv, GATConv
```

```python
# --- ノード埋め込み用エンコーダ（GCN） ---
class GCNEncoder(nn.Module):
    def __init__(self, in_dim, hidden_dim=32, out_dim=32):
        super().__init__()
        self.conv1 = GCNConv(in_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, out_dim)

    def forward(self, x, edge_index):
        h = self.conv1(x, edge_index)
        h = F.relu(h)
        z = self.conv2(h, edge_index)
        return z  # ノード埋め込み (N, out_dim)


# --- ノード埋め込み用エンコーダ（GAT） ---
class GATEncoder(nn.Module):
    def __init__(self, in_dim, hidden_dim=32, out_dim=32, heads=4):
        super().__init__()
        self.gat1 = GATConv(in_dim, hidden_dim, heads=heads, concat=True)
        self.gat2 = GATConv(hidden_dim * heads, out_dim, heads=1, concat=False)

    def forward(self, x, edge_index):
        h = self.gat1(x, edge_index)
        h = F.elu(h)
        z = self.gat2(h, edge_index)
        return z


# --- リンクスコアを出す層（内積ベース） ---
class LinkPredictor(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, z, edge_label_index):
        # edge_label_index: shape (2, E)
        src = z[edge_label_index[0]]   # (E, d)
        dst = z[edge_label_index[1]]   # (E, d)
        score = (src * dst).sum(dim=-1)  # 内積スコア
        return score  # シグモイドは外でかける（BCEWithLogitsLoss 用）
```

```python
from sklearn.metrics import roc_auc_score, average_precision_score
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("device:", device)

in_dim = data.num_node_features  # = 13 のはず

gcn_encoder = GCNEncoder(in_dim).to(device)
gat_encoder = GATEncoder(in_dim).to(device)
predictor_gcn = LinkPredictor().to(device)
predictor_gat = LinkPredictor().to(device)

train_data = train_data.to(device)
val_data = val_data.to(device)
test_data = test_data.to(device)

opt_gcn = torch.optim.Adam(
    list(gcn_encoder.parameters()) + list(predictor_gcn.parameters()),
    lr=0.01
)
opt_gat = torch.optim.Adam(
    list(gat_encoder.parameters()) + list(predictor_gat.parameters()),
    lr=0.01
)

criterion = torch.nn.BCEWithLogitsLoss()
```

```python
def train_epoch(encoder, predictor, data, optimizer):
    encoder.train()
    predictor.train()

    optimizer.zero_grad()

    z = encoder(data.x, data.edge_index)
    logits = predictor(z, data.edge_label_index)
    loss = criterion(logits, data.edge_label.float())

    loss.backward()
    optimizer.step()

    return float(loss.item())


@torch.no_grad()
def eval_link(encoder, predictor, data):
    encoder.eval()
    predictor.eval()

    z = encoder(data.x, data.edge_index)
    logits = predictor(z, data.edge_label_index)
    prob = torch.sigmoid(logits).cpu().numpy()
    y = data.edge_label.cpu().numpy()

    auc = roc_auc_score(y, prob)
    ap = average_precision_score(y, prob)
    return float(auc), float(ap)
```

```python
epochs = 200

for epoch in range(1, epochs + 1):
    loss_gcn = train_epoch(gcn_encoder, predictor_gcn, train_data, opt_gcn)
    loss_gat = train_epoch(gat_encoder, predictor_gat, train_data, opt_gat)

    if epoch % 20 == 0 or epoch == 1:
        val_auc_gcn, val_ap_gcn = eval_link(gcn_encoder, predictor_gcn, val_data)
        val_auc_gat, val_ap_gat = eval_link(gat_encoder, predictor_gat, val_data)
        print(f"[Epoch {epoch:03d}] "
              f"GCN loss={loss_gcn:.4f} val AUC={val_auc_gcn:.3f} AP={val_ap_gcn:.3f} | "
              f"GAT loss={loss_gat:.4f} val AUC={val_auc_gat:.3f} AP={val_ap_gat:.3f}")

# 最終テスト性能
test_auc_gcn, test_ap_gcn = eval_link(gcn_encoder, predictor_gcn, test_data)
test_auc_gat, test_ap_gat = eval_link(gat_encoder, predictor_gat, test_data)

print(f"GCN  test AUC={test_auc_gcn:.3f}, AP={test_ap_gcn:.3f}")
print(f"GAT  test AUC={test_auc_gat:.3f}, AP={test_ap_gat:.3f}")
```

```python
import numpy as np
import pandas as pd

@torch.no_grad()
def recommend_new_links(encoder, predictor, data_full, k=20):
    encoder.eval()
    predictor.eval()

    # 既存エッジ集合
    edge_set = set()
    ei = data_full.edge_index.cpu().numpy()
    for u, v in zip(ei[0], ei[1]):
        if u == v:
            continue
        a, b = sorted((int(u), int(v)))
        edge_set.add((a, b))

    # 未接続ペア候補を列挙
    N = data_full.num_nodes
    cand_pairs = []
    for i in range(N):
        for j in range(i + 1, N):
            if (i, j) not in edge_set:
                cand_pairs.append((i, j))

    edge_label_index = torch.tensor(cand_pairs, dtype=torch.long).t().to(device)

    # ノード埋め込み → ロジットスコア（内積）を計算
    z = encoder(data_full.x.to(device), data_full.edge_index.to(device))
    logits = predictor(z, edge_label_index)   # sigmoid はかけない
    scores = logits                           # そのままランキングに使う

    # Top-k を取得
    topk = min(k, scores.numel())
    vals, idx = torch.topk(scores, topk)

    cand_pairs = np.array(cand_pairs)
    top_pairs = cand_pairs[idx.cpu().numpy()]
    top_scores = vals.cpu().numpy()

    rows = []
    for (i, j), s in zip(top_pairs, top_scores):
        rows.append({
            "src_idx": int(i),
            "src_id": nodes.loc[i, "ID"],
            "src_name": nodes.loc[i, "企業名"],
            "src_sector": nodes.loc[i, "sector_group"],
            "dst_idx": int(j),
            "dst_id": nodes.loc[j, "ID"],
            "dst_name": nodes.loc[j, "企業名"],
            "dst_sector": nodes.loc[j, "sector_group"],
            "logit_score": float(s),     # ← ここがランキング指標
        })

    df_rec = pd.DataFrame(rows).sort_values("logit_score", ascending=False)
    return df_rec


# GAT で推薦リンク Top20
recommendations_gat = recommend_new_links(gat_encoder, predictor_gat, data, k=20)
print("=== GAT top-20 ===")
print(recommendations_gat.head(20))

# GCN 版も見たい場合
recommendations_gcn = recommend_new_links(gcn_encoder, predictor_gcn, data, k=20)
print("\n=== GCN top-20 ===")
print(recommendations_gcn.head(20))

# 表示精度を上げて logit の差を見る
pd.set_option("display.precision", 6)
print("\n=== GAT top-20 (with precision) ===")
print(recommendations_gat.head(20))
```

```python
import networkx as nx
import numpy as np
import pandas as pd
```

```python
def compute_trade_network_metrics(G: nx.Graph):
    """
    取引ネットワーク G について、よく使う 8 指標をまとめて計算する関数。
    ・ノード数 / エッジ数
    ・密度（density）
    ・最大連結成分サイズ（giant component）
    ・平均最短距離（average shortest path length, GC上）
    ・グローバル効率（global efficiency）
    ・平均クラスタリング係数（average clustering coefficient）
    ・次数分布の要約（平均・標準偏差）
    ・媒介中心性 / 固有ベクトル中心性の要約（平均・最大）
    """

    metrics = {}

    # --- 基本情報 ---
    metrics["num_nodes"] = G.number_of_nodes()
    metrics["num_edges"] = G.number_of_edges()
    metrics["density"] = nx.density(G)

    # --- 最大連結成分（Giant Component） ---
    if G.number_of_nodes() > 0:
        components = list(nx.connected_components(G))
        largest = max(components, key=len)
        Gc = G.subgraph(largest).copy()
        metrics["giant_component_size"] = len(largest)
    else:
        Gc = G.copy()
        metrics["giant_component_size"] = 0

    # --- 平均最短距離（GC 上で計算） ---
    try:
        if Gc.number_of_nodes() > 1:
            metrics["avg_shortest_path_length"] = nx.average_shortest_path_length(Gc)
        else:
            metrics["avg_shortest_path_length"] = None
    except Exception:
        metrics["avg_shortest_path_length"] = None

    # --- グローバル効率（全体） ---
    try:
        metrics["global_efficiency"] = nx.global_efficiency(G)
    except Exception:
        metrics["global_efficiency"] = None

    # --- 平均クラスタリング係数 ---
    try:
        metrics["avg_clustering"] = nx.average_clustering(G)
    except Exception:
        metrics["avg_clustering"] = None

    # --- 次数分布の要約 ---
    degrees = np.array([deg for _, deg in G.degree()])
    if len(degrees) > 0:
        metrics["degree_mean"] = float(degrees.mean())
        metrics["degree_std"] = float(degrees.std())
    else:
        metrics["degree_mean"] = None
        metrics["degree_std"] = None

    # --- 媒介中心性（betweenness） ---
    if G.number_of_nodes() > 0:
        bet = nx.betweenness_centrality(G, normalized=True)
        bet_vals = np.array(list(bet.values()))
        metrics["betweenness_mean"] = float(bet_vals.mean())
        metrics["betweenness_max"] = float(bet_vals.max())
    else:
        metrics["betweenness_mean"] = None
        metrics["betweenness_max"] = None

    # --- 固有ベクトル中心性（eigenvector centrality） ---
    try:
        eig = nx.eigenvector_centrality_numpy(G)  # 安定な方
        eig_vals = np.array(list(eig.values()))
        metrics["eigenvector_mean"] = float(eig_vals.mean())
        metrics["eigenvector_max"] = float(eig_vals.max())
    except Exception:
        metrics["eigenvector_mean"] = None
        metrics["eigenvector_max"] = None

    return metrics
```

```python
import itertools
import math
import numpy as np
import pandas as pd
import networkx as nx
```

```python
# ------------------------------------------------------------
# (A) Grab your base graph from existing variables (NO DUPES)
#     If your notebook already has a base graph, this will reuse it.
# ------------------------------------------------------------
def _pick_base_graph():
    # common names people use in the earlier part of the notebook
    candidates = ["G_base", "G0", "G", "base_G", "base_graph", "base_network"]
    for k in candidates:
        if k in globals() and isinstance(globals()[k], nx.Graph):
            return globals()[k]
    raise NameError(
        "Base graph not found. Please set one of these variables to your base networkx.Graph: "
        "G_base / G0 / G / base_G / base_graph / base_network"
    )

G0 = _pick_base_graph().copy()  # keep your original untouched
```

```python
# ------------------------------------------------------------
# (B) Metrics function (aligned to your Table 4-1 style)
#     Notes:
#       - avg_shortest_path_length: computed on the giant component (common choice when disconnected)
#       - global_efficiency: computed on full graph (nx.global_efficiency handles disconnected)
# ------------------------------------------------------------
def compute_metrics(G: nx.Graph) -> dict:
    # Giant component
    components = list(nx.connected_components(G))
    giant = max(components, key=len) if components else set()
    giant_size = len(giant)

    # avg shortest path length on giant component only
    if giant_size >= 2:
        Gg = G.subgraph(giant)
        asp = nx.average_shortest_path_length(Gg)
    else:
        asp = np.nan

    # global efficiency (full graph)
    ge = nx.global_efficiency(G)

    # clustering
    clust = nx.average_clustering(G)

    # degrees
    deg = np.array([d for _, d in G.degree()], dtype=float)
    deg_mean = float(deg.mean()) if len(deg) else 0.0
    deg_std = float(deg.std(ddof=0)) if len(deg) else 0.0

    # betweenness
    btw = nx.betweenness_centrality(G, normalized=True)
    btw_vals = np.array(list(btw.values()), dtype=float)
    btw_mean = float(btw_vals.mean()) if len(btw_vals) else 0.0
    btw_max = float(btw_vals.max()) if len(btw_vals) else 0.0

    # eigenvector centrality
    # (numpy method is stable for small graphs; for some graphs it can fail -> fall back to power method)
    try:
        eig = nx.eigenvector_centrality_numpy(G)
    except Exception:
        eig = nx.eigenvector_centrality(G, max_iter=5000, tol=1e-8)
    eig_vals = np.array(list(eig.values()), dtype=float)
    eig_mean = float(eig_vals.mean()) if len(eig_vals) else 0.0
    eig_max = float(eig_vals.max()) if len(eig_vals) else 0.0

    return {
        "giant_component_nodes": giant_size,
        "num_edges": G.number_of_edges(),
        "density": nx.density(G),
        "avg_shortest_path_length": asp,
        "global_efficiency": ge,
        "avg_clustering": clust,
        "degree_mean": deg_mean,
        "degree_std": deg_std,
        "betweenness_mean": btw_mean,
        "betweenness_max": btw_max,
        "eigenvector_mean": eig_mean,
        "eigenvector_max": eig_max,
    }

baseline = compute_metrics(G0)
```

```python
# ------------------------------------------------------------
# (C) Enumerate ALL missing edges (i<j, not already in E)
# ------------------------------------------------------------
nodes = list(G0.nodes())
existing = set((min(u, v), max(u, v)) for u, v in G0.edges())

missing_edges = [
    (u, v)
    for u, v in itertools.combinations(nodes, 2)
    if (min(u, v), max(u, v)) not in existing
]

print(f"nodes={len(nodes)}, edges={G0.number_of_edges()}, missing_edges={len(missing_edges)}")
```

```python
# ------------------------------------------------------------
# (D) Exhaustive evaluation
#     Saves: baseline metrics + new metrics + deltas for each edge
# ------------------------------------------------------------
rows = []
for (u, v) in missing_edges:
    G1 = G0.copy()
    G1.add_edge(u, v)

    m1 = compute_metrics(G1)
    delta = {f"delta_{k}": (m1[k] - baseline[k]) if (baseline[k] is not None and not (isinstance(baseline[k], float) and np.isnan(baseline[k]))) else np.nan
             for k in m1.keys()}

    rows.append({
        "u": u,
        "v": v,
        **m1,
        **delta,
    })

df_exhaustive = pd.DataFrame(rows)
```

```python
# ------------------------------------------------------------
# (E) Ranking examples (choose what you want)
#     1) Pure "inclusion": maximize giant component nodes
#     2) "Dispersion": minimize degree_std
#     3) "Bridge-friendly": reduce betweenness_max (avoid single-point bottleneck)
# ------------------------------------------------------------

# Example 1: inclusion-first
rank_inclusion = df_exhaustive.sort_values(
    by=["delta_giant_component_nodes", "delta_global_efficiency", "delta_degree_std"],
    ascending=[False, False, True]
)

# Example 2: dispersion-first (less concentration)
rank_dispersion = df_exhaustive.sort_values(
    by=["delta_degree_std", "delta_giant_component_nodes", "delta_global_efficiency"],
    ascending=[True, False, False]
)

# Example 3: balanced scalar score (keep metrics, just aggregate AFTER computing)
# (simple normalization by std to make units comparable)
score_cols = [
    "delta_giant_component_nodes",   # want +
    "delta_global_efficiency",       # want +
    "delta_avg_clustering",          # want + (but interpret carefully)
    "delta_degree_std",              # want -
    "delta_betweenness_max",         # want -
]
tmp = df_exhaustive[score_cols].copy()

# avoid zero std
stds = tmp.std(ddof=0).replace(0, 1.0)

df_exhaustive["score_balanced"] = (
      df_exhaustive["delta_giant_component_nodes"] / stds["delta_giant_component_nodes"]
    + df_exhaustive["delta_global_efficiency"]     / stds["delta_global_efficiency"]
    + df_exhaustive["delta_avg_clustering"]        / stds["delta_avg_clustering"]
    - df_exhaustive["delta_degree_std"]             / stds["delta_degree_std"]
    - df_exhaustive["delta_betweenness_max"]        / stds["delta_betweenness_max"]
)


rank_balanced = df_exhaustive.sort_values("score_balanced", ascending=False)
```

```python
# (F) Show top candidates  --- print() only, no display()

display_cols_basic = [
    "u", "v",
    "delta_giant_component_nodes",
    "delta_global_efficiency",
    "delta_avg_shortest_path_length",
    "delta_avg_clustering",
    "delta_degree_std",
    "delta_betweenness_max",
]

display_cols_with_score = display_cols_basic + ["score_balanced"]


def print_top(df, cols, title, n=15):

    print(title)
    print(df[cols].head(n).to_string(index=False))


print_top(
    rank_inclusion,
    display_cols_basic,
    "TOP 15 (inclusion-first)"
)

print_top(
    rank_dispersion,
    display_cols_basic,
    "TOP 15 (dispersion-first)"
)

print_top(
    rank_balanced,
    display_cols_with_score,
    "TOP 15 (balanced score)"
)
```

```python
# --- 1) 全探索結果の読み込み ---
exh = pd.read_excel(EXH_PATH)

# 必須列チェック
need_cols = {"u","v","delta_degree_std","delta_betweenness_max"}
missing = need_cols - set(exh.columns)
if missing:
    raise ValueError(f"Exhaustive results missing columns: {missing}")
```

```python
# --- 2) dispersion-firstスコア定義（シンプルで論文化しやすい） ---
#    目的：degree_std と betweenness_max を下げる（負方向が良い）
#    標準化して足し合わせ → 大きいほど良いにする
def zscore(s: pd.Series):
    s = s.astype(float)
    return (s - s.mean()) / (s.std(ddof=0) + 1e-12)

exh["z_delta_degree_std"] = zscore(exh["delta_degree_std"])
exh["z_delta_betweenness_max"] = zscore(exh["delta_betweenness_max"])

# dispersion-first: 下げたいのでマイナスを付けて "大きいほど良い"
exh["score_dispersion"] = -(exh["z_delta_degree_std"] + exh["z_delta_betweenness_max"])

# 「本当に集中緩和してるリンク」に絞る（両方が改善：<0）
exh_disp = exh[(exh["delta_degree_std"] < 0) & (exh["delta_betweenness_max"] < 0)].copy()

# 上位K
TOPK_IDEAL = 20
ideal_top = exh_disp.sort_values("score_dispersion", ascending=False).head(TOPK_IDEAL).copy()
```

```python
# --- 3) 企業属性テーブルを読み、u/vを企業名・業種に変換 ---
df_comp = pd.read_excel(COMP_PATH, sheet_name="Sheet1")

# ノードindexがdf_compの行番号(0-based)と一致している前提
NAME_COL = "企業名"
IND_COL  = "業種"

def idx_to_name(i: int) -> str:
    try:
        return str(df_comp.loc[int(i), NAME_COL])
    except Exception:
        return f"node_{i}"

def idx_to_ind(i: int) -> str:
    try:
        return str(df_comp.loc[int(i), IND_COL])
    except Exception:
        return ""

ideal_top["u_name"] = ideal_top["u"].apply(idx_to_name)
ideal_top["v_name"] = ideal_top["v"].apply(idx_to_name)
ideal_top["u_ind"]  = ideal_top["u"].apply(idx_to_ind)
ideal_top["v_ind"]  = ideal_top["v"].apply(idx_to_ind)

# 見やすい列順
cols_show = [
    "u","v","u_name","v_name","u_ind","v_ind",
    "delta_degree_std","delta_betweenness_max",
    "delta_global_efficiency","delta_avg_shortest_path_length",
    "score_dispersion"
]
cols_show = [c for c in cols_show if c in ideal_top.columns]
ideal_top = ideal_top[cols_show].copy()

print("=== dispersion-first 理想リンク TopK ===")
display(ideal_top)
```

```python
# --- 4) GAT/GCN推薦（recommendations_gat / recommendations_gcn）との重なりを見る ---
def to_undirected_set(df_pairs: pd.DataFrame, a="src_idx", b="dst_idx"):
    s = set()
    for _, r in df_pairs.iterrows():
        i, j = int(r[a]), int(r[b])
        if i == j:
            continue
        s.add(tuple(sorted((i, j))))
    return s

# recommendations_gat/gcn が notebook内に存在する前提（なければスキップ）
gat_set = set()
gcn_set = set()

try:
    gat_set = to_undirected_set(recommendations_gat, "src_idx", "dst_idx")
except Exception as e:
    print("[WARN] recommendations_gat not found or invalid:", e)

try:
    gcn_set = to_undirected_set(recommendations_gcn, "src_idx", "dst_idx")
except Exception as e:
    print("[WARN] recommendations_gcn not found or invalid:", e)

ideal_pairs = set(tuple(sorted((int(u), int(v)))) for u, v in ideal_top[["u","v"]].values)

ideal_top["in_GAT_top"] = [tuple(sorted((int(u), int(v)))) in gat_set for u, v in ideal_top[["u","v"]].values]
ideal_top["in_GCN_top"] = [tuple(sorted((int(u), int(v)))) in gcn_set for u, v in ideal_top[["u","v"]].values]

print("\n=== 理想リンクTopKがGAT/GCN推薦に入っているか ===")
display(ideal_top[["u","v","u_name","v_name","score_dispersion","in_GAT_top","in_GCN_top"]])

# ギャップ（理想なのに推薦されていない）
gap = ideal_top[(~ideal_top["in_GAT_top"]) & (~ideal_top["in_GCN_top"])].copy()
print("\n=== ギャップ（理想リンクなのにGAT/GCNどちらにも出ていない）===")
display(gap[["u","v","u_name","v_name","score_dispersion","delta_degree_std","delta_betweenness_max"]])

# 保存（卒論の表として使いやすい）
OUT_XLSX = "dispersion_first_gap_tables.xlsx"
ideal_top.to_excel(OUT_XLSX, index=False)
print(f"\nSaved: {OUT_XLSX}")
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# --- 0) 推薦DFにスコア列がある前提で logit を取る（無ければ順位だけでもOK） ---
def get_score_col(df):
    # notebookの作り方によって列名が違うので候補を順に探す
    for c in ["score", "logit", "prob", "p", "pred", "y_score"]:
        if c in df.columns:
            return c
    return None

# GAT/GCN の「元スコア」を辞書化（理想リンクの(logit)がないと介入の効果が見えないため）
def build_score_map(df, a="src_idx", b="dst_idx"):
    col = get_score_col(df)
    if col is None:
        print("[WARN] No score-like column found. Will use rank-based pseudo-logit.")
        # 代替：順位が高いほど大きい疑似logit
        df2 = df.copy().reset_index(drop=True)
        df2["__rank"] = np.arange(len(df2))
        df2["__pseudo_logit"] = (len(df2) - df2["__rank"]) / len(df2) * 6 - 3  # [-3, +3]くらい
        col = "__pseudo_logit"
    score_map = {}
    for _, r in df.iterrows():
        i, j = int(r[a]), int(r[b])
        key = tuple(sorted((i, j)))
        score_map[key] = float(r[col])
    return score_map

gat_score_map = {}
gcn_score_map = {}
try:
    gat_score_map = build_score_map(recommendations_gat, "src_idx", "dst_idx")
except Exception as e:
    print("[WARN] build_score_map for GAT failed:", e)

try:
    gcn_score_map = build_score_map(recommendations_gcn, "src_idx", "dst_idx")
except Exception as e:
    print("[WARN] build_score_map for GCN failed:", e)
```

```python
# --- 1) 理想リンク（ideal_top）に対して、現状logitと介入後logitを計算 ---
# どっちのモデルで介入を見る？（まずはGATで見るのが自然）
BASE_MODEL = "GAT"  # "GCN" でもOK
score_map = gat_score_map if BASE_MODEL.upper() == "GAT" else gcn_score_map

sim = ideal_top.copy()
sim["pair"] = [tuple(sorted((int(u), int(v)))) for u, v in sim[["u","v"]].values]
sim["base_logit"] = sim["pair"].map(score_map).fillna(np.nan)

# base_logitがNaNの場合：まだ推薦表に出ていない（=低い可能性）ので低い値を置く
# ここは論文的に「保守的に低め」へ
sim["base_logit"] = sim["base_logit"].fillna(-4.0)
```

```python
# --- 2) 政策介入の重み（手で調整するパラメータ） ---
w_match  = 1.5   # マッチング支援でlogitがどれだけ上がるか
w_credit = 1.0   # 信用補完でlogitがどれだけ上がるか

# 介入シナリオ（4通り）
scenarios = [
    ("none",   0, 0),
    ("match",  1, 0),
    ("credit", 0, 1),
    ("both",   1, 1),
]

for name, m, c in scenarios:
    sim[f"logit_{name}"] = sim["base_logit"] + w_match*m + w_credit*c
    sim[f"p_{name}"] = sigmoid(sim[f"logit_{name}"])
```

```python
# --- 3) 介入で「理想リンクがTopに上がる」かを見る（理想リンク集合内で順位比較） ---
rank_cols = []
for name,_,_ in scenarios:
    col = f"p_{name}"
    sim[f"rank_{name}"] = sim[col].rank(ascending=False, method="min")
    rank_cols.append(f"rank_{name}")

show_cols = ["u","v","u_name","v_name","score_dispersion","base_logit"] + \
            [f"p_{n}" for n,_,_ in scenarios] + rank_cols
show_cols = [c for c in show_cols if c in sim.columns]

print(f"=== 条件介入シミュレーション（{BASE_MODEL}ベース） ===")
display(sim[show_cols].sort_values("score_dispersion", ascending=False))
```

```python
# --- 4) 「最小介入」でTopに入るリンクを抽出（例：理想TopK内の上位5に入る等） ---
TOP_TARGET = 5
def min_intervention(row):
    # noneで入れば0
    if row["rank_none"] <= TOP_TARGET:
        return "none"
    if row["rank_match"] <= TOP_TARGET:
        return "match"
    if row["rank_credit"] <= TOP_TARGET:
        return "credit"
    if row["rank_both"] <= TOP_TARGET:
        return "both"
    return "hard"  # 何をしても上がらない（別の障壁/特徴が必要）

sim["min_intervention"] = sim.apply(min_intervention, axis=1)

print(f"=== 目標：理想リンク集合内でTop{TOP_TARGET}に入れるための最小介入 ===")
display(sim[["u","v","u_name","v_name","score_dispersion","min_intervention"]].sort_values("score_dispersion", ascending=False))

OUT_SIM = "dispersion_policy_intervention_sim.xlsx"
sim.to_excel(OUT_SIM, index=False)
print(f"Saved: {OUT_SIM}")
```
